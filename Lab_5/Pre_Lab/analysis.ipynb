```python
import numpy as np
import matplotlib.pyplot as plt

dark_35C = np.loadtxt('Dark/dark_35C.txt')
dark_45C = np.loadtxt('Dark/dark_45C.txt')
dark_55C = np.loadtxt('Dark/dark_55C.txt')
dark_65C = np.loadtxt('Dark/dark_65C.txt')
dark_75C = np.loadtxt('Dark/dark_75C.txt')

char_10_35C = np.loadtxt('Illuminated/char_10_35C.txt')
char_10_45C = np.loadtxt('Illuminated/char_10_45C.txt')
char_10_55C = np.loadtxt('Illuminated/char_10_55C.txt')
char_10_65C = np.loadtxt('Illuminated/char_10_65C.txt')
char_10_75C = np.loadtxt('Illuminated/char_10_75C.txt')
```

```python
V = dark_35C[:, 0] 
i_35 = dark_35C[:, 1] 
i_45 = dark_45C[:, 1] 
i_55 = dark_55C[:, 1]
i_65 = dark_65C[:, 1]
i_75 = dark_75C[:, 1]
plt.figure(figsize=(10, 6))


plt.plot(V, i_35, linestyle='-', label='dark_35C')
plt.plot(V, i_45,  linestyle='-', label='dark_45C')
plt.plot(V, i_55,  linestyle='-', label='dark_55C')
plt.plot(V, i_65,  linestyle='-', label='dark_65C')
plt.plot(V, i_75,  linestyle='-', label='dark_75C')

plt.xlabel('V')
plt.ylabel('I')
plt.title('dark IV characteristics at different temperatures')
plt.legend()
plt.show()
```

```python

i35_10 =-1* char_10_35C[:, 1] 
i45_10 = -1* char_10_45C[:, 1] 
i55_10 = -1* char_10_55C[:, 1] 
i65_10 = -1* char_10_65C[:, 1] 
i75_10 = -1* char_10_75C[:, 1] 

pi35_10 = -1* char_10_35C[:, 1]*V 
pi45_10 = -1 *char_10_45C[:, 1]*V 
pi55_10 = -1 * char_10_55C[:, 1]*V
pi65_10 = -1 * char_10_65C[:, 1]*V
pi75_10 = -1 * char_10_75C[:, 1]*V

plt.figure(figsize=(10, 8))

plt.plot(V, i35_10, linestyle='-', label='10mA 35C')
plt.plot(V, i45_10,  linestyle='-', label='10mA 45C')
plt.plot(V, i55_10,  linestyle='-', label='10mA 55C')
plt.plot(V, i65_10,  linestyle='-', label='10mA 65C')
plt.plot(V, i75_10,  linestyle='-', label='10mA 75C')

plt.plot(V, pi35_10, linestyle='-', label='power 10mA 35C')
plt.plot(V, pi45_10, linestyle='-', label='power 10mA 45C')
plt.plot(V, pi55_10, linestyle='-', label='power 10mA 55C')
plt.plot(V, pi65_10, linestyle='-', label='power 10mA 65C')
plt.plot(V, pi75_10, linestyle='-', label='power 10mA 75C')

# Labels and title
plt.xlabel('V')
plt.ylabel('I')
plt.title('light IV at diff temp')
plt.legend()
plt.show()
```

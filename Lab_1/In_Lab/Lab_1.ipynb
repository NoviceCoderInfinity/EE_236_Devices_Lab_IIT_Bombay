```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Your data
blue_V_d = pd.Series([3.04, 3.03, 2.98, 2.89, 2.88, 2.72, 2.54, 2.51, 2.25, 2.02, 1.74, 1.49, 1.11, 0.7, 0.63])
blue_I_d = pd.Series([9.3, 8.6, 6.5, 4.2, 3.11, 2.3, 1.9, 1.8, 1.5, 1.2, 1, 0.7, 0.4, 0.1, 1.00E-12])

# Convert pandas Series to numpy arrays
x = blue_V_d.values
y = blue_I_d.values

# Calculate the best fit line using numpy.polyfit
slope, intercept = np.polyfit(x, y, 1)

# Create the best fit line data
fit_line = slope * x + intercept

# Print the slope and intercept
print(f"Slope: {slope}")
print(f"Intercept: {intercept}")

# Plot the data and best fit line
plt.plot(x, y, 'b-', label='Blue LED Data')  # Data line
plt.scatter(x, y, color='blue', marker='x', label='Data Points')  # Data points
plt.plot(x, fit_line, 'r--', label='Best Fit Line')  # Best fit line
plt.xlabel('Voltage (V)')
plt.ylabel('Current (I)')
plt.title('Blue LED Current vs. Voltage with Best Fit Line')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy.stats import linregress

# Your data
blue_V_d = pd.Series([3.04, 3.03, 2.98, 2.89, 2.88, 2.72, 2.54, 2.51, 2.25, 2.02, 1.74, 1.49, 1.11, 0.7])
blue_I_d = pd.Series([9.3, 8.6, 6.5, 4.2, 3.11, 2.3, 1.9, 1.8, 1.5, 1.2, 1, 0.7, 0.4, 0.1])

# Convert pandas Series to numpy arrays
x = blue_V_d.values
y = blue_I_d.values

# Compute the logarithm of blue_I_d
# Add a small constant to avoid log(0) issues
y_log = np.log10(y + 1e-12)  # Use np.log10 for base-10 logarithm

# Perform linear regression
slope, intercept, r_value, p_value, std_err = linregress(x, y_log)

# Create the best fit line data
fit_line = slope * x + intercept

# Print the results
print(f"Slope: {slope}")
print(f"Intercept: {intercept}")
print(f"R-squared: {r_value**2}")
print(f"P-value: {p_value}")
print(f"Standard Error: {std_err}")

# Plot the data and best fit line
plt.plot(x, y_log, 'b-', label='Log of Blue LED Data')  # Data line
plt.scatter(x, y_log, color='blue', marker='x', label='Data Points')  # Data points
plt.plot(x, fit_line, 'r--', label='Best Fit Line')  # Best fit line
plt.xlabel('Voltage (V)')
plt.ylabel('Log10(Current)')
plt.title('Log of Blue LED Current vs. Voltage with Best Fit Line')
plt.legend()
plt.grid(True)
plt.show()
```

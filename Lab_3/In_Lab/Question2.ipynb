# Plot V_out v/s Wavelength for each intensity

## Intensity = 1000

```python
import matplotlib.pyplot as plt

# Data
wavelength = [950, 750, 520, 450]
v_out = [0.303, 0.24, 0.0332, 0.0139]
colors = ['red', 'red', 'green', 'blue']

# Plot
plt.figure(figsize=(8, 6))
plt.scatter(wavelength, v_out, color=colors, s=100)  # s is the size of the points
plt.plot(wavelength, v_out, linestyle='--', color='gray', alpha=0.5)  # Line to connect the points

# Labels and title
plt.xlabel('Wavelength (nm)', fontsize=12)
plt.ylabel('V_out (V)', fontsize=12)
plt.title('V_out vs Wavelength (Intensity = 1000)', fontsize=14)

# Display plot
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data for different intensities
wavelength = [950, 750, 520, 450]

v_out_1000 = [0.303, 0.24, 0.0332, 0.0139]
v_out_1500 = [0.359, 0.32, 0.0524, 0.0193]
v_out_2000 = [0.456, 0.38, 0.066, 0.0281]

colors = ['red', 'red', 'green', 'blue']

# Plot
plt.figure(figsize=(10, 8))

# Intensity = 1000
plt.scatter(wavelength, v_out_1000, color=colors, s=100, label='Intensity = 1000')
plt.plot(wavelength, v_out_1000, linestyle='--', color='gray', alpha=0.5)

# Intensity = 1500
plt.scatter(wavelength, v_out_1500, color=colors, s=100, marker='D', label='Intensity = 1500')
plt.plot(wavelength, v_out_1500, linestyle='--', color='gray', alpha=0.5)

# Intensity = 2000
plt.scatter(wavelength, v_out_2000, color=colors, s=100, marker='s', label='Intensity = 2000')
plt.plot(wavelength, v_out_2000, linestyle='--', color='gray', alpha=0.5)

# Labels and title
plt.xlabel('Wavelength (nm)', fontsize=12)
plt.ylabel('V_out (V)', fontsize=12)
plt.title('V_out vs Wavelength at Different Intensities', fontsize=14)

# Display legend
plt.legend()

# Display plot
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data for each LED
intensity = [1000, 1500, 2000]
v_out_red = [0.24, 0.32, 0.38]
v_out_green = [33.2, 52.4, 66.6]
v_out_blue = [13.9, 19.3, 28.1]
v_out_ir = [0.303, 0.359, 0.456]

# Plot
plt.figure(figsize=(10, 8))

# Red LED
plt.plot(intensity, v_out_red, marker='o', color='red', linestyle='-', label='Red LED (750 nm)')

# Green LED
plt.plot(intensity, v_out_green, marker='s', color='green', linestyle='-', label='Green LED (520 nm)')

# Blue LED
plt.plot(intensity, v_out_blue, marker='D', color='blue', linestyle='-', label='Blue LED (450 nm)')

# IR LED
plt.plot(intensity, v_out_ir, marker='^', color='purple', linestyle='-', label='IR LED (950 nm)')

# Labels and title
plt.xlabel('Intensity', fontsize=12)
plt.ylabel('V_out (V)', fontsize=12)
plt.title('V_out vs Intensity for Different LEDs', fontsize=14)

# Display legend
plt.legend()

# Display grid
plt.grid(True)

# Show plot
plt.show()
```
